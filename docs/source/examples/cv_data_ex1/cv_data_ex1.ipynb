```python
from dotenv import load_dotenv
import os
from pathlib import Path
load_dotenv()
print(os.getenv("PROJECT_FOLDER"))
projectPath = Path(os.getenv("PROJECT_FOLDER"))
print(projectPath)
```

```python
data = CV_Data(projectPath /"test_data" / "CV" / "CV_144700_ 3.tdms")
```

```python
import matplotlib.pyplot as plt
```

```python
data.plot()
plt.savefig("cv_data_ex1_fig1.png", dpi='figure', format=None)
```

```python
line, p = data.plot(dir="pos")
data.plot(dir="neg", plot = p)
plt.savefig("cv_data_ex1_fig2.png", dpi='figure', format=None)
```

```python
LSV_pos = data.get_sweep(POS) 
LSV_neg = data.get_sweep(NEG) 
LSV_avg = data.get_sweep(AVG) 
LSV_dif = data.get_sweep(DIF) 
```

```python
line,p = LSV_pos.plot(legend="pos")
LSV_neg.plot(plot=p,legend="neg")
LSV_avg.plot(plot=p,legend="avg")
LSV_dif.plot(plot=p,legend="dif")
```
