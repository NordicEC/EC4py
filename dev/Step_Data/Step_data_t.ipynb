```python
from ec4py import Step_Data, RHE, AREA, AREA_CM
import numpy as np
import copy
```

```python
data = Step_Data()
data.Time=np.array(range(0,100),dtype=np.float64)/10
data.E=data.Time
data.i=data.E*5
index = data.index_at_time(2.00)
print(index, data.Time[index])

```

```python

gdata = Step_Data()
size= 110
gdata.Time=np.array(range(0,size),dtype=np.float64)/20
gdata.E=np.ones(size,dtype=np.float64)
gdata.i=np.ones(size,dtype=np.float64)*2
```

```python
data = copy.deepcopy(gdata)
data.set_RHE(2)
data.area="4 m^2"
v = data.plot("i","Time", AREA)
data.set_RHE(1)
data.norm(AREA)
v2 = data.plot("i","Time")
print(v,v2)
```

Integrate Test

```python
 
data = copy.deepcopy(gdata)
data.set_RHE(2)
v = data.integrate(2.0,4.0,-1,RHE)
data.set_RHE(1)
v2 = data.integrate(2.0,4.0,-1,RHE)
print(v,v2)

```

```python
data = copy.deepcopy(gdata)
data.set_RHE(2)

data.area = "100 m^2"
v = data.integrate(2.0,4.0,-1, RHE,AREA)
data.norm("area")
v2 = data.integrate(2.0,4.0,-1, RHE)
print(v,v2)
```
